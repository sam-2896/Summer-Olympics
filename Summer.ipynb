# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. In how many cities Summer Olympics is held so far?

```python
df = pd.read_csv("summer.csv")
print(".................Till now " ,len(df["City"].unique())," cites organized olympics...................")
```

```python
print("....................top 5 cities are...................")

d =df[["City","Year"]].values.tolist()
a={}
l=[]
for x in d:
    if x[1] not in l:
        l.append(x[1])
        if x[0] in a:
            a[x[0]]+=1
        else:
            a[x[0]]=1
e =[]
for x in a:
    e.append([x,a[x]])
e = sorted(e, key=lambda x: x[1])[::-1]
c = [i[0] for i in e[:5:]]
b = [i[1] for i in e[:5:]]
fig = plt.figure(figsize = (10, 5))
plt.bar(c,b,color = "red",width = 0.4)
plt.ylabel("NO OF TIMES ")
plt.xlabel("CITY ")
plt.title("NO OF TIMES CITY HOLD OLYMPICS ( BAR GRAPHS)")
plt.legend(["no of time olympic held"],loc="upper right" , facecolor = "white",edgecolor = "red",fontsize = 10, shadow  = True, )
plt.grid(b = True, color ='black',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
plt.show()
```

```python
colors = ['lightskyblue', 'red', 'blue', 'green', 'gold'] 
fig = plt.figure(figsize =(10, 7))
plt.pie(b,labels=b, shadow=True)
plt.legend(c,title ="City",
          loc ="center right",bbox_to_anchor =(1, 0, 0.5, 1))
plt.show()
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
a=[]
b={}
v=[]
c= df[["Sport","Medal"]].values.tolist()

for t,y in c :
    if y=="Gold":
        if t in b.keys():
            b[t]=b[t]+1
        else :
            b[t]=1
for g in b:
    a.append([g,b[g]])
aa = sorted(a, key=lambda x: x[1])[::-1]
e = [i[0] for i in aa[:5:]]
f = [i[1] for i in aa[:5:]]
for a,b in aa[:5:]:
    print(f"          #   '{a}'   with  {b}   GOLD Medals")
```
